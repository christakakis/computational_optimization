```python
A = ([[   100,  -50,  50,    20],
       [   10, 5000,  25, -2000],
       [  200, -100, 100,   -50],
       [ 1000,  250, -50,  -100]])

b = [100, 200, 300, 400]

c = [50, -100, 50, -25]

A = np.float_(A)
b = np.float_(b)
c = np.float_(c)
```
